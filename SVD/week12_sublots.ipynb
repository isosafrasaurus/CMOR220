```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x  = np.linspace(0,1)
y  = x*x
fig, axes = plt.subplots(nrows=1, ncols=6)
for i in range(6):
    axes[i].plot(x,y)
```

```python
x  = np.linspace(0,1)
y  = x-1
fig, axes = plt.subplots(nrows=2, ncols=3)
axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[0,2].plot(x,y)
axes[1,0].plot(x,y+1)
axes[1,1].plot(x,y+2)
axes[1,2].plot(x,y+3)
```
